### Cleaning Data

```python
day_df['mnth'] = day_df['mnth'].map({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
    7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'})

day_df['season'] = day_df['season'].map({1: 'Springer', 2: 'Summer', 3: 'Fall', 4: 'Winter'})

day_df['yr'] = day_df['yr'].map({0: 2011, 1: 2012})

day_df['weekday'] = day_df['weekday'].map({0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'})

day_df['weathersit'] = day_df['weathersit'].map({1: 'Clear, Few clouds, Partly cloudy, Partly cloudy', 2: 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist',
                                                   3: 'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds',
                                                   4: 'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'})

day_df['workingday'] = day_df['workingday'].map({0: 'Holiday', 1: 'Workingday'})

day_df['mnth'] = day_df.mnth.astype('category')
day_df['season'] = day_df.season.astype('category')
day_df['yr'] = day_df.yr.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['weathersit'] = day_df.weathersit.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df
```

```python
hour_df['mnth'] = hour_df['mnth'].map({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
    7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'})

hour_df['season'] = hour_df['season'].map({1: 'Springer', 2: 'Summer', 3: 'Fall', 4: 'Winter'})

hour_df['yr'] = hour_df['yr'].map({0: 2011, 1: 2012})

hour_df['weekday'] = hour_df['weekday'].map({0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'})

hour_df['weathersit'] = hour_df['weathersit'].map({1: 'Clear, Few clouds, Partly cloudy, Partly cloudy', 2: 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist',
                                                   3: 'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds',
                                                   4: 'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'})

hour_df['workingday'] = hour_df['workingday'].map({0: 'Holiday', 1: 'Workingday'})

hour_df['mnth'] = hour_df.mnth.astype('category')
hour_df['season'] = hour_df.season.astype('category')
hour_df['yr'] = hour_df.yr.astype('category')
hour_df['holiday'] = hour_df.holiday.astype('category')
hour_df['weekday'] = hour_df.weekday.astype('category')
hour_df['weathersit'] = hour_df.weathersit.astype('category')
hour_df['workingday'] = hour_df.workingday.astype('category')
hour_df
```

```python
day_df.to_csv("day_cleaned.csv", index=False)
hour_df.to_csv("hour_cleaned.csv", index=False)
```

## Exploratory Data Analysis (EDA)

### Explore `hour_df`

Pola tren penyewaan sepeda harian (Perjam)

---



```python
trend_bike_hourly = hour_df.groupby(by=["hr", "weekday"]).agg({
    'cnt': 'sum'
})

trend_bike_hourly
```

```python
trend_bike_hourly = hour_df.groupby(by=["hr","weekday"]).agg({
    "cnt": "sum"
}).reset_index()

weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(13, 5))
sns.lineplot(
    data=trend_bike_hourly,
    x="hr",
    y="cnt",
    hue="weekday",
    hue_order=weekday_order,
    palette="viridis",
    marker="o")

plt.title("Pola Sewa Sepeda Harian (Perjam)")
plt.xlabel("Jam")
plt.ylabel("Jumlah Pengguna Sepeda")
plt.legend()
plt.xticks(np.arange(0, 24, 1))
plt.tight_layout()
plt.show()
```

Pada grafik terlihat bahwa terdapat 2 pola yang berbeda yaitu untuk Workingday dan Holiday. Pada Workingday (Senin s.d. Jumat) terjadi peningkatan jumlah penyewaan sepeda pada jam 8.00 dan juga jam 17.00-18.00. Sedangkan pada Holiday (Sabtu dan Minggu) puncak jumlah penyewaan sepeda terjadi pada jam 12.00-16.00.

Pola tren penyewaan sepeda tiap musim (Perjam)


```python
trend_bike_hourly = hour_df.groupby(by=["hr", "season"]).agg({
    'cnt': 'sum'
})

trend_bike_hourly
```

```python
season_trend_hourly = hour_df.groupby(by=["hr","season"]).agg({
    "cnt": "sum"
}).reset_index()

plt.figure(figsize=(13, 5))
sns.lineplot(
    data=season_trend_hourly,
    x="hr",
    y="cnt",
    hue="season",
    palette="rocket",
    marker="o")

plt.title("Pola Sewa Sepeda Tiap Musim (Perjam)")
plt.xlabel("Jam")
plt.ylabel("Jumlah Pengguna Sepeda")
plt.legend()
plt.xticks(np.arange(0, 24, 1))
plt.tight_layout()
plt.show()
```

Pola tren penyewaan sepeda untuk tiap musim relatif mirip yaitu terjadi puncak jumlah penyewaan sepeda pada jam 8.00 dan jam 17.00-18.00. Dimana musim Springer menjadi musim dengan jumlah pengguna sepeda paling sedikit dan musim Fall menjadi musim dengan jumlah pengguna sepeda paling banyak.

### Explore `day_df`

Tren penyewaan sepeda tiap tahunnya

```python
day_df['mnth'] = pd.Categorical(day_df['mnth'], categories=
    ['January','February','March','April','May','June','July','August','September','October','November','December'],
    ordered=True)

trend_bike_daily = day_df.groupby(by=["mnth", "yr"]).agg({
    'cnt': 'sum'
})

trend_bike_daily
```

```python
trend_bike_daily = day_df.groupby(by=["mnth","yr"]).agg({
    "cnt": "sum"
}).reset_index()

plt.figure(figsize=(11, 5))
sns.lineplot(
    data=trend_bike_daily,
    x="mnth",
    y="cnt",
    hue="yr",
    palette="viridis",
    marker="o")

plt.title("Tren Sewa Sepeda")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Pengguna Sepeda")
plt.legend(title="Tahun")
plt.tight_layout()
plt.show()
```

Tren sewa sepeda meningkat ditahun 2012 jika dibandingkan dengan tahun 2011. Terjadi penurunan pada pertengahan tahun 2011 hingga akhir tahun. Kemudian meningkat drastis di tahun 2012.

Perbandingan penyewaan sepeda harian

```python
weekday_trend_daily = day_df.groupby(by='weekday').agg({
    'casual': 'sum',
    'registered': 'sum',
    'cnt':['max','min','mean','sum']
}).reindex (['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])

weekday_trend_daily
```

```python
day_df['weekday'] = pd.Categorical(day_df['weekday'], categories=
    ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'],
    ordered=True)

weekday_trend_daily = day_df.groupby('weekday')[['registered', 'casual']].sum().reset_index()

plt.figure(figsize=(10,6))

plt.bar(
    weekday_trend_daily['weekday'],
    weekday_trend_daily['registered'],
    label='Registered',
    color='#97BC62FF'
)

plt.bar(
    weekday_trend_daily['weekday'],
    weekday_trend_daily['casual'],
    label='Casual',
    color='#2C5F2D'
)

plt.title('Perbandingan Penyewa Sepeda Setiap Hari')
plt.xlabel('Hari')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.legend()
plt.show()
```

Terjadi penurunan penyewaan sepeda oleh pengguna pada hari Sabtu dan Minggu. Jumlah penyewaan sepeda meningkat lagi pada hari Senin hingga Jumat. Pengguna lebih jarang menyewa sepeda pada weekend dibandingkan weekday (Senin s.d. Jumat).
